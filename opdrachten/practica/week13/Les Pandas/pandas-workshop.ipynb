Wat is dan de prijs voor een pot Nutella in elk jaar?

```python
# Voeg een kolom 'Nutella' toe aan het DataFrame:



def productionprize(df, comp,totalw):
    totprice = pd.Series(data=[0]*df.shape[0])
    for k in comp:
        perc = comp[k]
        cpricepkg = df[k]
        #totprice+=cpricepkg*totalw*perc/100
        temp = cpricepkg*totalw*perc/100
        totprice+=temp
    return totprice
nutella   = {'Cocoa':50,
             'Sugar':20,
             'Maize':5,
             'Coconut Oil':20}

totprice = productionprize(df,nutella,0.8)
print(totprice)

# ...

# Bereken de prijs voor een pot van 800 gram over de jaren heen:

# ...
```
